## Test OMOP query functions on GEMINI (gemini_omop_beta2) DB

```python
import cyclops.query.gemini_omop as gemini
from cyclops.query.interface import QueryInterface
```

```python
tables = [table for table in dir(gemini._db.ods) if "__" not in table]
tables
```

## Visit occurrence table

```python
visit_occurrence_table = QueryInterface(gemini._db, gemini._db.ods.visit_occurrence)
visit_occurrence_table.run(limit=1000)
print(visit_occurrence_table.data.columns)
visit_occurrence_table.data
```

```python
visits_query_interface = gemini.visit_occurrence(years=[2019], months=[1, 4])
visits = visits_query_interface.run()
visits
```
